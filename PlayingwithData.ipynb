```python
import numpy as np
import mne
import matplotlib.pyplot as plt
```

```python
raw_data = mne.io.read_raw_eeglab('data/ds004504/derivatives/sub-002/eeg/sub-002_task-eyesclosed_eeg.set')
```

The above command reads the EEG of one of the participant and stores in raw_data. As you can see below, it has information about all the different channels

```python
print(raw_data.info)
```

The following command splits the data for each channel into 4s epochs and then puts the time series into a Pandas dataframe. 

```python
epochs = mne.make_fixed_length_epochs(raw_data, duration = 4.0)
dfsub_002 = epochs.to_data_frame()
dfsub_002.head(5)
```
